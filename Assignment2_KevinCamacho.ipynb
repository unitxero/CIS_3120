```python
#in-class exercise


import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Baruch_College'
r = requests.get(url)
soup = BeautifulSoup(r.text)


href_list = []

for item in soup.select('li a[href]'):
  href_list.append(item.get('href'))

for item in href_list:
  print(item)
```
